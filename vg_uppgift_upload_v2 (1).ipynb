## 1.1 Kvantiler

Kvantiler är skärpunkter som delar intervallet för en sannolikhetsfördelning i kontinuerliga intervall med lika sannolikheter, eller delar observationerna i ett urval på samma sätt. Det finns en kvantil mindre än antalet skapade grupper. Vanliga skärningspunkter är kvartiler, deciler, och percentiler.

q-kvantiler är värden som delar upp en ändlig uppsättning värden i q delmängder av (nästan) lika stora. Det finns q − 1 partitioner av q-kvantilerna, en för varje heltal k som uppfyller 0 < k < q. I vissa fall kanske värdet av en kvantil inte är unikt bestämt, vilket kan vara fallet för medianen (2-kvantil) för en enhetlig sannolikhetsfördelning på en uppsättning av jämn storlek. Kvantiler kan också tillämpas på kontinuerliga distributioner, vilket ger ett sätt att generalisera rankningsstatistik till kontinuerliga variabler (se percentilrankning). När den kumulativa fördelningsfunktionen för en slumpvariabel är känd är q-kvantilerna tillämpningen av kvantilfunktionen (den inversa funktionen av den kumulativa fördelningsfunktionen) på värdena {1/q, 2/q, …, (q − 1)/q}.

## Below we have three plots (Density plot, Histogram and QQ-plot) for the normal distribution, uniform distribution and exponential distribution respectively. This code is for you to more easily grasp the concept and feel free to play around with the code for your own learning experience. 

```r
# Case: Normal Distribution. 
par(mfrow = c(1, 3))

x <- seq(from = -5, to = 5, by = 0.01)
y1 <- rnorm(1000, mean = 0 , sd = 1)
y2 <- dnorm(x, mean = 0 , sd = 1)

plot(x, y2, type="l", main = 'Density Plot')
hist(y1)
qqnorm(y1)
qqline(y1)
```

```r
# Case: Uniform Distribution. 
par(mfrow = c(1, 3))

x <- seq(from = -1, to = 2, by = 0.01)
y1 = dunif(x, min = 0, max = 1)
y2 <- runif(80, min = 0, max = 1)

plot(x, y1, type="l", main = 'Density Plot')
hist(y2)
qqnorm(y2)
qqline(y2)
```

```r
# Case: Exponential Distribution.
par(mfrow = c(1, 3))

x <- seq(from = 0, to = 10, by = 0.01)
y1 = dexp(x, rate = 1)
y2 <- rexp(80, rate = 1)

plot(x, y1, type="l", main = 'Density Plot')
hist(y2)
qqnorm(y2)
qqline(y2)
```

## 3. Shapiro-Wilks Normality Test

In the shapiro-Wilk test, the null hypothesis is: 

$H_{0}$ = data is normally distributed. 


Hence, if the p-value is greater than 0.05, we cannot reject that the data is normally distributed. If the p-value is smaller than 0.05, we reject the null hypothesis meaning we do not "believe" the data is normally distributed. Of course, you can choose whatever significance level you wish, here I assumed $\alpha = 0.05$ which is always the case if nothing else is said. 

As you can read in Wikipedia (https://en.wikipedia.org/wiki/Shapiro%E2%80%93Wilk_test): *\"(...)if the sample size is sufficiently large this test may detect even trivial departures from the null hypothesis\"* so using normal hypothesis tests should not be done "without thinking".

Kopplade till test är nollhypoteser som bestämmer mer specifikt hur fördelningen av testvariabeln ser ut. # 5.1 Confidence Intervals are a very useful tool in every data scientists toolbox. A colleague asks you - what is a confidence interval? Explain it to him/her! 

Konfidensintervall är en uppskattning av osäkerhet som är associerad med skattningar av populationsparameterar som genererats med hjälp av stickprovsdata. Konfidensintervall genereras med given konfidensgrad. 

Låt oss säga ett en för en population så vill du skatta en förutbestämd parameter utifrån stickprovsdata. Du tar flera prover och en intervallskattning för parametern bestäms. Konfidensintervallet är det intervall som innesluter populationsparametern. Det innebär att om man skulle upprepa undersökningen så skulle vi i 95% täcka in det sanna (men okända) värdet. Om konfidensgrad är 95% så kommer populations parameterna innelsuta 95% utav proverna. Ett ensidigt konfidensintervall begränsar populationen från ett håll. Ett tvåsidigt från både ner och upp. 


Med ett sannolikhetsurval kan man beräkna det värde man vill skatta inte bara som ett enda värde utan också med ett intervall s.k. konfidensintervall. Konfidensintervallets storlek påverkas av observationer i urvalet och spridning för variabeln.
Är skillnadens storlek viktig? Problem kan kräva en mer detaljerad mindre felmarginal och ibland är det inte så viktigt hur stor osäkerheten är, utan man vill mer titta bredare. Om skillnaden ligger inom felmarginalen så betyder det att den är så liten att den ryms inom osäkerheten.

För konfidensintervall gäller att:

- Större/mindre standard avikelse ger kortare/längre intervall.
- Flera/färre observationer n ger kortare/längre intervall.
- Större/mindre konfidensgrad 1 - $\alpha$ ger kortare/länger intervall.
